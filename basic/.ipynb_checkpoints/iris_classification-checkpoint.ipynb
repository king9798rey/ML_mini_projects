```python
from sklearn.linear_model import LogisticRegression
```

```python
y_pred=LR.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score , f1_score , confusion_matrix , recall_score, precision_score
```

```python
print("accuracy_score is ",accuracy_score(y_pred,y_test))
```

```python
print("precision_score is ", precision_score(y_pred,y_test, average="macro"))
```

```python
print("confusion_matrix is ",confusion_matrix(y_pred,y_test))
```

```python
print("f1_score is ",f1_score(y_pred,y_test,average="macro"))
```
